1장. 딥러닝과 파이토치 라이브러리 소개
================================

1.1 딥러닝 혁명
----------

> 머신러닝
 - 과거에 피처 엔지니어링에 크게 의존
 - 피처 엔지니어링: 다운스트림 알고리즘이 새로운 데이터를 기반으로 올바를 결과물을 낼 수 있도록 촉진하기 위해 데이터를 변환하는 작업
 - 다운스트림 알고리즘이 문제를 풀기 위해 올바를 변환 규칙을 포함해야 함
 - 예) 0과 1을 구분하기 위해 테두리선, 또는 원의 갯수 등을 사용
> 딥 러닝
 - 머신러닝과 다르게, 원본 데이터로부터 숫자의 규칙성을 찾아냄 --> 자동으로 개선
 - 딥 러닝이 부상하지만 피처 엔지니어링 또한 사전 지식을 주입하기 위해 사용됨
 - 때로는 딥 러닝보다 피처 엔지니어링을 사용했을 때 결과물이 좋을때도 있다
> 피처 vs 딥
 - 피처: 실무자가 피처를 잘 만들어 낼 수록 좋은 결과 나옴
 - 딥: 원본 데이터가 바로 들어가며 성능 최적화의 유도과정에 따라 저수준에서 고수준의 피처들이 자동으로 추출
> 딥러닝을 성공시키기 위한 필요사항
 - 가지고 있는 데이터가 무엇이든 일단 넣을 수 있어야 한다.
 - 딥러닝 머신을 정의해야한다.
 - 유용한 표현을 얻고 머신이 원하는 출력 결과를 낼 수 있도록 자동화된 훈련 방법을 만들어야 한다.
> 훈련이란?
 - 기대한값과 실제 모델의 차이를 비교하기 위한 참조 데이터, 모델의 출력값을 사용한 실수 함수로 기준을 만듦
 - 기준을 통해 더 낮은 점수를 얻을 수 있도록 딥러닝 머신을 변경하는 과정
 - 결과적으로 기존에 입력되지 않은 데이터에 대해서도 낮은 점수를 받도록 함

1.2 딥러닝을 위한 파이토치
-------------

> 파이토치란?
 - 딥러닝 프로젝트를 위한 파이썬 프로그램 라이브러리
 - 딥러닝 모델을 표현할 수 있다.
 - 가장 각광받는 딥러닝 도구
> 파이토치에 대하여
 - 'tensor'라는 핵심 데이터 구조를 제공하며, 이를 사용하여 복잡한 수학 함수를 만들 수 있다.
 - tensor: 다차원 배열로, 자료구조 위에서 전용 하드웨어를 활용하여 수학 연산을 수행한다.
 - 벡터, 행렬, 배열을 담을 수 있는 데이터 타입
 - 따라서 신경망 구조를 설계하고 개별/분산 컴퓨팅 자원에서 훈련하기 매우 편리하다.

1.3 왜 파이토치인가?
----------------

> 파이토치를 사용하는 이유
 - 단순함: 학습, 활용, 확장, 디버깅 면에서 모두 수월함 (파이썬과 유사)
 - CPU에서 처리하는 것보다 GPU를 활용하면 50배나 빠른 결과물을 도출할 수 있다.
 - 과학 계산에서도 유리하게 사용될 수 있다.
 - 불필요한 복잡함 없이 정교하게 모델을 구현할 수 있다.
 - 파이썬에 의존하지 않고, C++을 사용할 수도 있다.

1.4 파이토치 딥러닝 프로젝트 둘러보기
-----------------------------

> 파이토치의 활용성
 - 텐서: 다양한 연산 라이브러리를 torch 모듈로 제공 
 - CPU에서 수행하던 연산을 GPU로 쉽게 옮길 수 있다.
 - 특정 텐서에서 수행한 모든 연산을 기억했다가 주어진 입력값을 기준으로 미분값을 자동 제공해줌
> 파이토치 프로젝트의 기본 구조
 1. 저장소 등 데이터 출처로부터 물리적으로 데이터를 가져오기
 2. 데이터에서 가져온 샘플을 파이토치가 다룰 수 있는 텐서로 변환 (torch.utils.data에 있는 Dataset class)
 3. 텐서를 batch 형태로 묶어야 하는데, 이때 병렬 처리로 진행한다.(DataLoader class에 존재)
 4. 훈련 루프로 넘어간다. 이때 대부분 파이썬의 for 문을 사용한다. (출력값을 기대값과 비교, 기준/손실함수 사용)
 - 훈련 루프는 딥러닝 프로젝트에서 가장 많은 시간을 소요하는 부분이다.
 5. 훈련이 끝나면 훈련된 모델이 만들어진다.
 6. 모델이 실제 작업이 필요한 곳에서 활용되도록 해야한다. 그 후 배포 진행
